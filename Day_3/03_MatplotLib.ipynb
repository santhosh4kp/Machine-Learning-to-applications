```python
import numpy as np
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('angle')
plt.ylabel('sin')
plt.title('sin wave')
plt.show()
```

```python
plt.plot(x,np.sin(x), 'b+', label='sin')
plt.plot(x,np.cos(x) ,'y--', label='cos') #Scatter connections
plt.xlabel('angle')
plt.ylabel('sin/cos')
plt.title('sin/cos wave')
plt.ylim(-2,2)
plt.xlim(-5,15)
plt.legend()
plt.show()
```

## Subplot

```python
fig, axis = plt.subplots(1,2, figsize=(10,5))
print(axis.shape)
```

```python
fig, axis = plt.subplots(1,2, figsize=(15,5))
x = np.arange(0,10,0.1)
axis[0].plot(x,np.sin(x), 'g--')
axis[0].set_title('sin')
axis[0].set_xlabel('angle')
axis[0].set_ylabel('sin')
axis[1].plot(x,np.cos(x), 'r--')
axis[1].set_title('cos')
axis[1].set_xlabel('angle')
axis[1].set_ylabel('cos')
plt.show()
```

```python
fig, axis = plt.subplots(2,2, figsize=(10,10))
x = np.arange(0,10,0.1)
axis[0][0].plot(x,np.sin(x), 'y--')
axis[0][0].set_title('sin')
axis[0][0].set_ylim(-3,3)
axis[0][1].plot(x,2*np.sin(x), 'g--')
axis[0][1].set_title('sin')
axis[0][1].set_ylim(-3,3)
axis[1][0].plot(x,np.cos(x), 'b--')
axis[1][0].set_title('cos')
axis[1][0].set_ylim(-3,3)
axis[1][1].plot(x,2*np.cos(x), 'r--')
axis[1][1].set_title('cos')
axis[1][1].set_ylim(-3,3)
plt.show()
```

```python
df = pd.read_csv('/content/marketing_customer_data_uncleaned.csv')
df.columns

```

```python
# prompt: show the number of people with differnt group in marital status

print(df['Marital_Status'].value_counts())
```

```python

sns.countplot(x = 'Marital_Status', data = df)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Marital_Status', palette='viridis')

# Add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Count of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```
