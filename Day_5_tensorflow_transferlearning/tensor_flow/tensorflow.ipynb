```python
#from tensorflow import keras
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Input,Dense,Add,Concatenate

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :]  # we only take the first two features.
y = iris.target
print(y)
encoder=LabelBinarizer()
Y=encoder.fit_transform(y)
print(X)
print(Y)
```

```python
# type 1

model = Sequential()
model.add(Dense(100, activation='relu', input_shape = (4,)))
model.add(Dense(50, activation='relu'))
model.add(Dense(3, activation='softmax'))

#model = Model(input, output )


model.compile(optimizer='sgd',loss="categorical_crossentropy",metrics=['accuracy'])

model.summary()

tf.keras.utils.plot_model(model,show_shapes=True,  dpi=70)
```

```python
# functional
input = Input(shape=(4,))
x1 = Dense(100, activation='relu')(input)
x21 = Dense(50, activation='relu')(x1)
x22 = Dense(50, activation='relu')(x1)
#ut_shape = (4,)))
# model.add(Dense(50, activation='relu'))
# model.add(Dense(3, activation='softmax'))
x2 = Add()([x21,x22])
x3 = Concatenate()([x1,x2])
output = Dense(3, activation='softmax')(x3)
model = Model(input, output )


model.compile(optimizer='sgd',loss="categorical_crossentropy",metrics=['accuracy'])

model.summary()

tf.keras.utils.plot_model(model,show_shapes=True, dpi=50)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model.fit(X_train,y_train,epochs=5,batch_size=16)
model.evaluate(X_test,y_test)
```

```python
y_pred  = np.argmax(model(X_test), axis =-1)
print(y_pred)

y_true =  np.argmax(y_test, axis =-1)
print(y_true)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_true, y_pred)
print(cm)
cr = classification_report(y_true, y_pred)
print(cr)
```

```python
import seaborn as sns
sns.heatmap(cm,cmap='Greens', annot=True,xticklabels=iris.target_names,yticklabels=iris.target_names)
```
