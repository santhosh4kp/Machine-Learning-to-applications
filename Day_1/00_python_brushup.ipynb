```python
# prime number generation

limit = int(input('Enter number : '))
for num in range(2,limit+1):
    is_divisible = False
    k=2
    while k <= num//2 :
        if num % k == 0:
            is_divisible=True
            break;
        k += 1   #k=k+1
    if not is_divisible:
        print(num,end=' ')
```

**Object Oriented Programming**

Class and Objects

```python
class Adder:

    #constructor
    def __init__(self):
        self.__x = 0
        self.__y = 0
    #setter
    def setValues(self, x, y):
        self.__x = x
        self.__y = y

    #self variable name can be anything
    def calculate(self1):
        self1.__sum = self1.__x + self1.__y

    #getter
    def getSum(self):
        return self.__sum
```

```python
adder =  Adder()
adder.calculate()
print(adder.getSum())
```

```python

a = Adder()
a.setValues(10, 20)
a.calculate()
print(a.getSum())  # Output: 30
```
