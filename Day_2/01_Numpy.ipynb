```python
# prompt: aritematic expression in string  evaluation by input

import numpy as np

def evaluate_expression(expression):
    """Evaluates an arithmetic expression provided as a string.

    Args:
    expression: The arithmetic expression to evaluate.

    Returns:
    The result of the evaluation, or an error message if the expression is invalid.
    """
    try:
        result = eval(expression)
        return result
    except (SyntaxError, NameError, TypeError):
        return "Invalid expression"


if __name__ == "__main__":
    expression = input("Enter an arithmetic expression: ")
    result = evaluate_expression(expression)
    print(f"Result: {result}")
```

```python
# prompt: store the details of employees and show it in ascending order of their name

employee_details = []

while True:
    name = input("Enter employee name (or type 'done' to finish): ")
    if name.lower() == 'done':
        break

    try:
        age = int(input("Enter employee age: "))
        salary = float(input("Enter employee salary: "))
    except ValueError:
        print("Invalid input. Please enter numeric values for age and salary.")
        continue

    employee_details.append([name, age, salary])


employee_details.sort()

print("\nEmployee Details (Sorted by Name):")
for emp in employee_details:
    print(f"Name: {emp[0]}, Age: {emp[1]}, Salary: {emp[2]}")
print(employee_details)
```
